#### Data Analysis & Visulization

```python
cancellation_percentage = df["is_canceled"].value_counts(normalize = True)
cancellation_percentage
```

```python
plt.figure(figsize = (4,5))
plt.title("Reservation Status")
plt.bar(['Not Canceled','Canceled'],df["is_canceled"].value_counts())
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot (x = 'hotel', hue ='is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels ()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel("reservations")
plt.legend(['not canceled','canceled'])
plt.show()
```

#### It is cleared that City hotel have more cancellations

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (10,8))
plt.title("Average Daily Rate in City and Resort hotel", size = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,18))
ax_1 = sns.countplot(x = 'month' , hue = 'is_canceled', data = df , palette = 'bright')
legend_labels,_ = ax_1. get_legend_handles_labels ()
ax_1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('Adr per Montn' , fontsize = 20)
sns.barplot('month','adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_countries = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title("Top 10 countries with Canceled Reservations", fontsize =  15)
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countires.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

non_canceled_data = df[df['is_canceled']== 0]
non_canceled_df_adr = non_canceled_data.groupby('reservation_status_date')[['adr']].mean()
non_canceled_df_adr.reset_index(inplace = True)
non_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rates')
plt.plot(non_canceled_df_adr['reservation_status_date'],non_canceled_df_adr['adr'], label = 'not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'canceled')
plt.legend()
```
